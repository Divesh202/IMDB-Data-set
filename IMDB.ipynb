```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("/content/imdb_top_1000.csv")
df.tail(3)
```

```python
df.head(3)
```

```python
df.info()
```

```python
df.drop('Poster_Link',axis=1,inplace=True)
```

##`Change The Data Type`

```python
df["Released_Year"].unique()
```

```python
# Released_year convert the object to int
df=df[df['Released_Year']!="PG"]
df["Released_Year"] = df['Released_Year'].astype(int)
```

```python
df["Gross"] = (df["Gross"]
.str.replace("$","",regex = False)
.str.replace(",","")
.astype(float))
```

```python
df.info()
```

##`Check The Duplicate Value`

```python
# No Any Duplicate Values
df.duplicated().sum()
```

#`Check The Null Values`

```python
df.isnull().sum()
```

```python
df['Certificate'] = df['Certificate'].fillna('Unrated')
df['Meta_score'].fillna(df['Meta_score'].mode()[0], inplace=True)
df['Gross'].fillna(df['Gross'].mode()[0], inplace=True)

print(df.isnull().sum())
```

```python
df['Stars'] = df[['Star1', 'Star2', 'Star3', 'Star4']].apply(lambda x: ','.join(x.dropna()), axis=1)
df = df.drop(columns = ['Star1', 'Star2', 'Star3', 'Star4'])
df.head(5)
```

## **`Most Popular Genre`**

```python
Top_5 = df['Genre'].value_counts().nlargest(5)
Top_5
```

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=Top_5.index, y=Top_5.values , color = "#FF33A1")
ax.bar_label(ax.containers[0])
plt.title("Top 5 Genres Distribution")
plt.xticks(rotation=45)
plt.show()
```

## **`Top rated movies`**

```python
top_rated_movies = df.sort_values(by='IMDB_Rating', ascending=False).head(10)
top_rated_movies = top_rated_movies[['Series_Title', 'IMDB_Rating']]
top_rated_movies = pd.DataFrame(top_rated_movies).reset_index(drop = True)
top_rated_movies.index += 1
top_rated_movies
```

## **`Top director by number of movies`**

```python
top_directors = df['Director'].value_counts().nlargest(10)
plt.figure(figsize=(10,6))
ax = sns.barplot(x=top_directors.values, y=top_directors.index, color = "#FF33A1")
ax.bar_label(ax.containers[0])
plt.title("Top Directors by Number of Movies")
plt.show()
```

## **`Most earned movies as percentage of total revenue`**

```python
top_movies = df.sort_values(by='Gross', ascending=False)
top_movies = top_movies[['Series_Title', 'Gross']].head(5)
top_movies = pd.DataFrame(top_movies).reset_index(drop=True)

top_movies['Gross_Percentage'] = (top_movies['Gross'] / top_movies['Gross'].sum()) * 100
```

```python
plt.figure(figsize=(10, 6))
ax = sns.barplot(data=top_movies, x='Series_Title', y='Gross_Percentage',color = '#FF33A1')

for p in ax.patches:
    ax.annotate(f'{p.get_height():.1f}%',
                (p.get_x() + p.get_width() / 2, p.get_height()),
                ha='center', va='bottom', fontsize=10, color='black')

plt.title("Most Earned Movies (Percentage)")
plt.ylabel("Gross Revenue (%)")
plt.xlabel("Movies")
plt.xticks(rotation=90)

plt.show()
```
